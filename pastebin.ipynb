```python
# pull out the strings (could have done this more simply with split I think)
import re
hosts = re.findall('[a-z0-9]*\.[a-z0-9\.]*', hostnames)
```

```python
hosts
```

```python
# look each of them up, keep track of the set of them, print the match
import socket
ips = set()
for host in hosts:
    try:
        ip = socket.gethostbyname(host)
        ips.add(ip)
        print('{} -> {}'.format(host,ip))
    except:
        print('No ip address for {}'.format(host))
        
# summarize what we found
print('Collected {} unique ips in {} domain names'.format(len(ips), len(hosts)))
```
